```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter, WeekdayLocator,DayLocator, AutoDateLocator
%matplotlib inline
```

```python
df = pd.read_csv('curtailments.csv')
```

```python
plt.plot(df['Wind and solar curtailment'])
plt.title('Figure 1: California solar and wind monthly curtailments, 2014 - YTD 2018')
plt.ylabel('Curtailed output in Megawatt-hours')
plt.show()
```

**Figure 1** shows how much wind and solar electricity output was curtailed (reduced) on a monthly basis in California since mid 2014. Curtailments occur because too much electricity is being generated when it is not needed. Curtailments spike on a periodic basis, likely because of seasonal mismatch between supply and demand (in the spring and fall when demand is low and solar output is relatively high). Note that curtailments increase over time - this is because more wind and solar power plants have been added to the California grid over the past few years.

_Note_: I had trouble formatting the x-axis to show date intervals which is why dates are not shown

```python
plt.title('Figure 2: Level and frequency of monthly curtailments of wind and solar in California, 2014 - YTD 2018')
plt.ylabel('Frequency (number of months)')
plt.xlabel('Monthly curtailment level in Megawatt-hours')
plt.hist(df['Wind and solar curtailment'], color = 'gold', bins = 20)

plt.show()
```

**Figure 2** shows that over the last few years, the level of wind and solar curtailments in California has generally been below 20,000 Megawatt-hours in any given month, or (very) roughly around 1% or less of total wind/solar ouptut. Because the number of wind and solar power plants has been growing rapidly in California over the past couple of years, it would be interesting to compare the same plot for 2014-2016 and 2017-2018.

```python
first_2_years = df.loc[0:27]
last_2_years = df.loc[27:53]

plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
plt.hist(first_2_years['Wind and solar curtailment'], color = 'green')
plt.ylim([0,10])
plt.xlim([0,95000])
plt.title('Figure 3a. Curtailments from 5/2014-8/2016')

plt.subplot(1, 2, 2)
plt.hist(last_2_years['Wind and solar curtailment'], color = 'red')
plt.ylim([0,10])
plt.xlim([0,95000])
plt.title('Figure 3b. Curtailments from 8/2016-10/2018')
plt.show()
```

**Figure 3a** and **Figure 3b** show the level and frequency of solar and wind monthly curtailments between May 2014 and August 2016 and August 2016 and November 2018, respectively. Early on, as Figure 3a shows, the majority of monthly curtailments were well below 20,000 MWhs, with a few months around 40,000 MWhs. Over the past couple of years, a similar pattern exists except that it has shifted to the right, clustering around the 20-40K MWh level at the low end and 80K MWh level on the high end.
