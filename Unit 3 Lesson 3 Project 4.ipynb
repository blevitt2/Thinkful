```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lognormalvar = np.random.lognormal(1,0.5,1000)
gammavar = np.random.gamma(2,2,1000)
exponentvar = np.random.exponential(2,1000)
poissonvar= np.random.poisson(2,1000)
standard_normalvar = np.random.standard_normal(1000)
weibullvar = np.random.weibull(2,1000)
```

```python
plt.hist(lognormalvar, bins=20)
plt.axvline(lognormalvar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(lognormalvar.mean() + lognormalvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(lognormalvar.mean()- lognormalvar.std(), color='b', linestyle='dashed', linewidth=2) 
plt.title('lognormal')
plt.show()
```

In general, the mean and std dev are a good indicators of where the data tends to cluster and how closely they cluster together, especially in distributions that resemble the normal distribution. The mean and std dev are not as good at indicating central tendency in skewed distributions like the lognormal.,<br /><br />

Alone, the mean and std dev do not indicate: If and how the data skews (left / right), whether the data has more than one point of central tendency, if and where there are clusters of outliers, if there is no central tendency, etc.  

```python
normalvar1 = np.random.normal(5,0.5,1000)
normalvar2 = np.random.normal(10,1,1000)
normalvar3 = normalvar1 + normalvar2   
```

```python
plt.hist(normalvar3, bins=20)
plt.axvline(normalvar3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(normalvar3.mean() + normalvar3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(normalvar3.mean()- normalvar3.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
```

```python
print(normalvar3.mean())
print(normalvar3.std())
```

The mean is roughly the same as the addition of the means of the two underlying distributions. However, the standard deviation of the new distribution is not equal to the sum of the two standard deviations of the underlying distributions -- it is in between the two underlying standard deviations.
